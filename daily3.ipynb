```python
import unittest

def has_right_options(position: str) -> bool:
    n = len(position)

    for i in range(n):
        if position[i] == 'F':
            if i - 1 >= 0 and position[i - 1] == ' ':
                return True

            if i - 2 >= 0 and position[i - 1] == 'T' and position[i - 2] == ' ':
                return True

    return False


def get_right_options(position: str) -> list[str]:
    options = []
    n = len(position)

    for i in range(n):
        if position[i] == 'F':
            if i - 1 >= 0 and position[i - 1] == ' ':
                chars = list(position)
                chars[i], chars[i - 1] = chars[i - 1], chars[i]
                options.append(''.join(chars))

            if i - 2 >= 0 and position[i - 1] == 'T' and position[i - 2] == ' ':
                chars = list(position)
                chars[i], chars[i - 2] = chars[i - 2], chars[i]
                options.append(''.join(chars))

    return options


def get_Right_options(position: str) -> list[str]:
    return get_right_options(position)


class TestDailyThree(unittest.TestCase):

    def test_has_right_options_slide(self):
        self.assertTrue(has_right_options(" FT"))

    def test_has_right_options_jump(self):
        self.assertTrue(has_right_options(" TF"))

    def test_has_right_options_none(self):
        self.assertFalse(has_right_options("FTF"))

    def test_get_right_options_slide(self):
        self.assertEqual(get_right_options(" FT"), ["F T"])

    def test_get_right_options_jump(self):
        self.assertEqual(get_right_options(" TF"), ["FT "])

    def test_get_right_options_multiple(self):
        self.assertEqual(get_right_options(" F TF"), ["F  TF", " FFT "])


if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
