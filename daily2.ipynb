```python
import unittest

def has_left_options(position: str) -> bool:
    n = len(position)

    for i in range(n):
        if position[i] == 'T':
            if i + 1 < n and position[i + 1] == ' ':
                return True

            if i + 2 < n and position[i + 1] == 'F' and position[i + 2] == ' ':
                return True

    return False


def get_left_options(position: str) -> list[str]:
    options = []
    n = len(position)

    for i in range(n):
        if position[i] == 'T':
            if i + 1 < n and position[i + 1] == ' ':
                chars = list(position)
                chars[i], chars[i + 1] = chars[i + 1], chars[i]
                options.append(''.join(chars))

            if i + 2 < n and position[i + 1] == 'F' and position[i + 2] == ' ':
                chars = list(position)
                chars[i], chars[i + 2] = chars[i + 2], chars[i]
                options.append(''.join(chars))

    return options


class TestDailyTwo(unittest.TestCase):

    def test_has_left_options_slide(self):
        self.assertTrue(has_left_options("T F"))

    def test_has_left_options_jump(self):
        self.assertTrue(has_left_options("TF "))

    def test_has_left_options_none(self):
        self.assertFalse(has_left_options("FTF"))

    def test_get_left_options_slide(self):
        self.assertEqual(get_left_options("T F"), [" TF"])

    def test_get_left_options_jump(self):
        self.assertEqual(get_left_options("TF "), [" FT"])

    def test_get_left_options_multiple(self):
        self.assertEqual(get_left_options("T TF "), [" TTF ", "T  FT"])


if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
